```python
#Import dependencies
import pandas as pd
import os
```

```python
student_file_to_load=os.path.join("Resources","Students_complete.csv")
student_data_df=pd.read_csv(student_file_to_load)
student_data_df.head()
```

```python
#Put the students name in a list
student_names=student_data_df["student_name"].tolist()
```

```python
#List students with incorrect names
students_to_fix=[]
#Loop through the student names
for name in student_names:
    #Check if the student has more than 2 names.
    if len(name.split())>=3:
        #add the name to the list
        students_to_fix.append(name)
#print list length
len(students_to_fix)
```

```python
#Create a list of prefixes
prefixes=[]
#Loop trough all the names to fix
for name in students_to_fix:
    #Check if the first name has 4 or less letters
    if len(name.split()[0])<=4:
        #Add the name to the list
        prefixes.append(name.split()[0])
#Get the unique items of the list
set(prefixes)
```

```python
#Create a list to hold the suffixes
suffixes=[]
#loop through all the names
for name in students_to_fix:
    #Check if the last name has 3 letters or less
    if len(name.split()[-1])<=3:
        #add the name to the lest
        suffixes.append(name.split()[-1])
#Get the unique items of the list
set(suffixes)
```

```python
#Create a list with the most common prefixes and suffixes
prefixes_suffixes=["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
#Conver the column into a str instead of an object
#loop through the prefixes_suffixes list
for word in prefixes_suffixes:
    student_data_df["student_name"]=student_data_df["student_name"].str.replace(word,"")
student_data_df.head(10)
```

```python
student_names=student_data_df["student_name"].tolist()
students_fixed=[]
for name in student_names:
    if len(name.split())>=3:
        students_fixed.append(name)
len(students_fixed)
```
